```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import numpy as np # linear algebra
import matplotlib 
import matplotlib.pyplot as plt
import sklearn
%matplotlib inline 
```

```python
df = pd.read_csv("gases.csv")
```

```python
df.head()
```

```python
graph = pd.pivot_table(df, values='value', index=['country_or_area', 'year'], columns=['category'])
graph.head()
```

```python
graph.plot()
```

```python
gases = graph.columns.values
```

```python
def plotCountry(nameOfCountry):
    data = graph.loc[nameOfCountry]
    plt.plot(data)
    plt.legend(gases)
    plt.title(nameOfCountry)
```

```python
plotCountry('Finland') 
```

```python
def plotSpecificGas(nameOfGas): 
    graph.plot( y = nameOfGas) 
    
plotSpecificGas(gases[7]) #gases[7] is no2 without landuse and forestry
```

```python
table = pd.pivot_table(df, values='value', index=['category', 'year'], columns=['country_or_area'])
table.head()
```

```python
def plotSpecificGas(nameOfGas): 
    graph.plot( y = nameOfGas) 
    
plotSpecificGas(gases[0]) #gases[0] is co2 without landuse and forestry
```

```python
countryNames = table.columns.values
```

```python
def gas_countries_plot(nameOfGas, namesOfCountries):
    data = table.loc[nameOfGas]
    data.plot( y = namesOfCountries)
    plt.legend(namesOfCountries)
    plt.title(nameOfGas)
gas_countries_plot(gases[0],countryNames[:10]) #plots the co2 levels for the first 10 countries on the list
```

```python
importantCountries = ['United States of America','Finland','Russian Federation', 'Germany', 'Australia',   'Sweden',]

gas_countries_plot(gases[0],importantCountries) #compares these countries amongst each other for co2
```

```python
gas_countries_plot(gases[7],importantCountries) #compares these countries amongst each other for no2
```
